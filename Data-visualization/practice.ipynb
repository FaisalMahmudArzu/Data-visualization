```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("All working. Miracle.")
```

Matplotlib basics

```python
x = np.linspace(0,5,11)
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X axis")
```

Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

Object Oriented Way

```python
fig = plt.figure(figsize=(12,8),dpi=500)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.3,0.4,0.4])
axis2.plot(x,y)
```

Saving Plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("Basicplot.png")
```

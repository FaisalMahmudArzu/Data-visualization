```python
import seaborn as sns
import matplotlib.pyplot as plt
```

We use distribution plots to visualize the distribution of quantitative data.
Let's see some plots that allow us to visualize the distribution of a data set. These plots are:
Displot/Hisplot
jointplot
pairplot
rugplot

Displot/Hisplot

```python
df = sns.load_dataset("tips")
df
```

```python
df["size"].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df["total_bill"],bins = 20,kde = True)
plt.subplot(1,2,2)
sns.histplot(df["tip"],bins = 20,kde = True)
# KDE - Kernel Density Estimate
```

Jointplot

```python
sns.jointplot(x = "total_bill",y = "tip",data = df,kind = "scatter")
```

```python
sns.jointplot(x = "total_bill",y = "tip",data = df,kind = "kde")
```

```python
sns.jointplot(x = "total_bill",y = "tip",data = df,kind = "hex")
```

```python
sns.jointplot(x = "total_bill",y = "tip",data = df,kind = "reg")
```

Pairplot

```python
sns.pairplot(df,hue='size',palette='rainbow')
# hue categorizes the data based on the column provided. In this case, it will categorize the data based on the "sex" column, which contains the values
```

```python
sns.rugplot(df)
```
